```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##Read Data Set
```

```python
data = pd.read_csv(r'C:\Users\AKHIL\Downloads\archive (5)\movies.csv')
```

```python
data.head()
```

```python
## Filling of null values with mean
```

```python
df = data.fillna(data.mean())
```

```python
df.head()
```

```python
## Know the Data Type
```

```python
df.dtypes
```

```python
## Change Datatype to Different Columns
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
## Change the format of the date column
```

```python
df['format_released'] = pd.to_datetime(df['released'].str.extract(r'(\w+ \d+, \d+)', expand=False), format='%B %d, %Y')
```

```python
df['Exact Year'] = df['format_released'].astype(str).str[0:4]
```

```python
df.head()
```

```python
## Sorting of gross column values in desecnding order
```

```python
df = df.sort_values(by = ['gross'],inplace = False,ascending = False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
## Droping of null values
```

```python
df.drop_duplicates()
df.head()
```

```python
## Arrange all the numeric value to all the columns eg:- country column has a usa,india like wise but we arrange a numeric value like usa 1, india 2 . we maintain that categeory witha numeric value.
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
#
```

```python
## Plot heatmap to know the corr values of the categeory using numeric values
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot =True)
```

```python
## usnig of the correlation matrix unstack the matrix like pairs . every colulmn has the own correlation.
```

```python
correlation_mat = df_numerized.corr(method = 'pearson')
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
## Arrange the pairs in sorting order with abpve 0.5 high correlation
```

```python
sorted_pairs=corr_pairs.sort_values()
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
```
